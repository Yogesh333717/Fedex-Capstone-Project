## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Fill missing categorical values with 'Unknown'
categorical_cols = df.select_dtypes(include= 'object').columns
df[categorical_cols] = df[categorical_cols].fillna("Unknown")

# Fill missing numerical values with 0 or median
numerical_cols = df.select_dtypes(include = ['int64', 'float64']).columns
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].median())

# As there is some values which are related to Date/time we convert them to datetime
date_cols = [col for col in df.columns if "Date" in col]
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
#Convert numeric columns with mixed types
df['Freight Cost (USD)'] = pd.to_numeric(df['Freight Cost (USD)'], errors='coerce')

#Add new column
df['Delivery_Delay_Days'] = (df['Delivered to Client Date'] - df['Scheduled Delivery Date']).dt.days
df['Delivery_Year'] = df['Delivered to Client Date'].dt.year
df['Delivery_Month'] = df['Delivered to Client Date'].dt.month

# Avoid false data like negative entries
df = df[df['Line Item Quantity'] > 0]
df = df[df['Unit Price'] >= 0]
````
